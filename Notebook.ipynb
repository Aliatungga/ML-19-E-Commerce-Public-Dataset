### Pertanyaan 2:

#### Performa Transaksi bulanan

#### Transaksi Trends

```python
trans_month = sell.groupby(['month'])['payment_value'].agg(['count','sum']).reset_index()
trans_month
```

```python
# trends transaction
def plot_trans_month(trans_month, x, y, title="", xlabel='month', ylabel='Number of buy transaction', dpi=100):
    plt.figure(figsize=(50,10), dpi=dpi)
    plt.plot(x, y, color='tab:blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()


plot_trans_month(trans_month, x=trans_month['month'], y=trans_month['count'], title='Number of buy transaction')
```

**INSIGHT**

Trends transaksi cenderung fluktuatif. Namun, pernah terjadi penurunan transaksi yang cukup drastis pada bulan september 2016, oktober 2016, desember 2016, september 2018 dan september 2018

```python
stat_order = sell.groupby(by="order_status").order_id.nunique().sort_values(ascending=False).reset_index()
stat_order
```

```python
fig = plt.figure(figsize = (15,7))

fig.suptitle('status order', fontsize=20)
plt.xlabel('order_id', fontsize=20)
plt.ylabel('order_status', fontsize=20)

total_data = sell['customer_id'].count()

stat_order.sort_values('order_id', ascending=True, inplace=True)
x = stat_order['order_status']
y = stat_order['order_id']

plt.barh(x, y, color=('lightblue'))
plt.show()
```

**INSIGHT**

Dari banyaknya transaksi yang telah dilakukan, sebanyak 98.478 transaksi yang berhasil deliverd. Namun, terdapat 2 transaksi yang hanya berhenti di approved dan 5 transaksi berhenti di created.

```python
delivery_time = sell['order_delivered_customer_date'] - sell['order_delivered_carrier_date']
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
sell['delivery_time'] = round(delivery_time/86400)
```

```python
sell['delivery_time'].describe()
```

**INSIGHT**

Rata-Rata waktu pengirman yang dilakukan mencapai 9 hari. Dengan pengiriman paling cepat 4 hari dan paling lama 12 hari.
